```python
# import dependencies
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
# get the html file using request
url = 'https://www.vlr.gg/91583/akrew-vs-100-thieves-knights-circuit-monthly-2022-april-ro16/?game=all&tab=overview'
html_txt = requests.get(url)
soup = BeautifulSoup(html_txt.text, 'lxml')
# get the match_stats
game_stats = soup.find('div', class_="vm-stats-container")
table_info = game_stats.find('table', class_='wf-table-inset mod-overview')
```

```python
# create a game_id variable
game_id = []
number_of_games = 1

i = 1
while i <= number_of_games:
    x = 0
    while x < 5:
        game_id.append(i)
        x+=1
    i+=1
```

```python
# get the map
# map to the area we need
stat_game = soup.find('div', class_='vm-stats-game')
game_header = stat_game.find('div', class_='vm-stats-game-header')
div_map = game_header.find('div', class_='map')
div_style = div_map.find('div')
# get the text needed
span_map = div_style.find('span').text

# remove the weird letters
p = '\n\t*'
p1 = '\t*'
span_map = re.sub(p, '', span_map)
span_map = re.sub(p1, '', span_map)

# remove the word PICK from the string
i = []
for x in span_map:
    i.append(x)
i.remove('P')
i.remove('I')
i.remove('C')
i.remove('K')

# append the map list so that it is equal length 
map = []
map_str = ''.join(i)

i = 0
while i<5:
    map.append(map_str)
    i+=1

print(map)
```

```python
# score for team L
# map to the data
stat_game = soup.find('div', class_='vm-stats-game')
game_header = stat_game.find('div', class_='vm-stats-game-header')
div_team = game_header.find('div', class_='team')

# create lists to hold the data
score_left = []
ct_score_left = []
t_score_left = []
team_left = []

# append the values to each list
score_l = div_team.find('div', class_='score').text
ct_score_l = div_team.find('span', class_='mod-ct').text
t_score_l = div_team.find('span', class_='mod-t').text
team_l = div_team.find('div', class_='team-name').text

# remove the weird text by using regex
p = '\n\t*'
p1 = '\t*'
team_l = re.sub(p, '', team_l)
team_l = re.sub(p1, '', team_l)

# append the lists to have 5 of each in them so that we can paste it into the table
i = 0
while i<5:
    team_left.append(team_l)
    score_left.append(int(score_l))
    ct_score_left.append(int(ct_score_l))
    t_score_left.append(int(t_score_l))
    i+=1

print(team_left)
print(score_left)
print(ct_score_left)
print(t_score_left)
```

```python
# score for team R
stat_game = soup.find('div', class_='vm-stats-game')
game_header = stat_game.find('div', class_='vm-stats-game-header')
div_team_r = game_header.find('div', class_='team mod-right')

# create lists to hold the data
score_right = []
ct_score_right = []
t_score_right = []
team_right = []

# append the values to each list
score_r = div_team_r.find('div', class_='score').text
ct_score_r = div_team_r.find('span', class_='mod-ct').text
t_score_r = div_team_r.find('span', class_='mod-t').text
team_r = div_team_r.find('div', class_='team-name').text

# remove the weird text by using regex
p = '\n\t*'
p1 = '\t*'
team_r = re.sub(p, '', team_r)
team_r = re.sub(p1, '', team_r)

# append the lists to have 5 of each in them so that we can paste it into the table
i = 0
while i<5:
    team_right.append(team_r)
    score_right.append(int(score_r))
    ct_score_right.append(int(ct_score_r))
    t_score_right.append(int(t_score_r))
    i+=1

print(team_right)
print(score_right)
print(ct_score_right)
print(t_score_right)
```

```python
# get all the player names and team abreviation
body = table_info.find('tbody')
tr = body.find_all('tr')

# create a variable to hold our webscraped info
a = []

# use a while loop to access each players info and send it into the list
x=0
while x < len(tr):
    trx = tr[x]
    td_name = trx.find('td', class_='mod-player')
    div = td_name.find('div')
    a.append(div.find('a'))
    x+=1

# create lists to hold each string of information
name_web=[]
team_web=[]

# use a while loop to find the info that we need and put it into the lists
i=0
while i < len(a):
    ai = a[i]
    name_web.append(ai.find('div', class_='text-of').text)
    team_web.append(ai.find('div', class_='ge-text-light').text)
    i+=1

# create our final variables to hold the player information
name = []
team = []
team_close = []
p = '\n\t*'
p1 = '\t*'
for n in name_web:
    name.append(re.sub(p, '', n))


for n in team_web:
    team_close.append(re.sub(p, '', n))
for b in team_close:
    team.append(re.sub(p1, '', b))
print(name)
print(team)
```

```python
# mapping to the correct location for all player for acs, kills, deaths, etc.
body = table_info.find('tbody')
tr = body.find_all('tr')

# create lists to hold the data
acs_both = []
acs_t = []
acs_ct = []

# use a while loop to get to the information for each player and add it into our lists
x=0
while x < len(tr):
    trx = tr[x]
    td_stat = trx.find('td', class_='mod-stat')
    span_sq = td_stat.find('span', class_='stats-sq')
    acs_both.append(span_sq.find('span', class_='side mod-side mod-both').text)
    acs_t.append(span_sq.find('span', class_='side mod-side mod-t').text)
    acs_ct.append(span_sq.find('span', class_='side mod-side mod-ct').text)
    x+=1

# set all the list data into integers
acs_both = list(map(int, acs_both))
acs_t = list(map(int, acs_t))
acs_ct = list(map(int, acs_ct))

# print to double check I am right
print(acs_both)
print(acs_t)
print(acs_ct)
```
